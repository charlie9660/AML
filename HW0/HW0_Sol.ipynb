```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import urllib
import pandas as pd
from StringIO import StringIO
```

```python
#1. Data Loading
link = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
f = urllib.urlopen(link)
myfile = StringIO(f.read())
attributes = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width(cm)","species"] 
df = pd.read_csv(myfile,sep = ",",names = attributes)
print(df['species'].value_counts())
```

```python
#2. Data Parsing
attribute_array = df.as_matrix(columns = attributes[:4])
label_vector = df['species'].values.T.tolist()
print(attribute_array)
print(label_vector)
```

```python
#3. Data Plotting
import seaborn as sns
sns.pairplot(df, hue="species")
#pd.scatter_matrix(df[["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width(cm)"]],figsize=(10, 10),c=['r','b','g'],)
plt.show()
```
